```python
#numpy exercise 2
import numpy as np
arr = np.array([12.5, 15.2, 14.8, 13.5, 16.0, 17.1, 15.6])
print(arr)
```

```python
total = np.sum(arr)
average = np.mean(arr)
print(total)
print(average)
```

```python
minimum = np.min(arr)
maximum = np.max(arr)
diff = np.diff(arr)
print(
    "Minimum Temp : \n", minimum,
    "\n Maximum Temp : \n",maximum,
    "\n Difference between temperature every day : \n", diff
)
```

```python
print(arr[arr>16])
```

```python
#numpy exercise 3
import numpy as np
marks = np.array([85, 90, 78, 88, 92])
Highest = np.max(marks)
Lowest = np.min(marks)
Average = np.mean(marks)
print(Highest,Lowest,Average)
```

```python
difference = marks - Average
print(difference)
```

```python
percentage = marks.astype(float)
print(percentage)
```

```python
find = marks[marks>80]
print(find)
```

```python
#numpy exercise 4
import numpy as np
stock = np.array([102.5, 104.2, 101.8, 103.0, 105.5, 107.3])
print(stock)
print(" Maximum = ", np.max(stock),"\n",
      "Minimum = ", np.min(stock),"\n",
      "Average = ", np.mean(stock)
)
```

```python
diff = np.diff(stock)
print(diff)
```

```python
percentage_change = (diff / stock[:-1]) * 100
percentage_change = np.round(percentage_change, 2)
print(percentage_change)
increased = percentage_change[percentage_change>1.5]
print(increased)
```
