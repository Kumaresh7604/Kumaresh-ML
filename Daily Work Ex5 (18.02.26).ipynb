```python
#Daily Exercise of Student DB
#Numpy Exercise 5
import numpy as np
np.random.seed(50)
marks = np.random.randint(10, 101,size =(10,5))
print(marks)
```

```python
average_by_sub = np.mean(marks,axis=0)
print("Average =",average_by_sub)
Highest_mark_by_sub = np.max(marks,axis=0)
print("Highest Mark per Sub =", Highest_mark_by_sub)
Lowest_mark_by_sub = np.min(marks,axis=0)
print("Lowest Mark per Sub =", Lowest_mark_by_sub)
```

```python
Class_Topper = np.sum(marks,axis = 1)
print("The Topper of the Class is",np.argmax(Class_Topper),"with highest total of",np.max(Class_Topper),"out of 500")
```

```python
Pass_Sub = marks>35
print(Pass_Sub)
Pass_count = np.sum(Pass_Sub, axis = 0)
print(Pass_count)
```

```python
difficult = np.argmin(Pass_count)
print(difficult)
Lowest_Avg = np.min(average_by_sub)
print(np.min(average_by_sub))
```
