```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    data = { "Product": ["Laptop", "Phone", "Tablet", "Camera", "Headset"],
    "Mon": [5, 10, 4, 2, 8],
    "Tue": [6, 12, 5, 3, 9],
    "Wed": [3, 11, 2, 1, 7],
    "Thu": [4, 9, 3, 2, 6],
    "Fri": [7, 13, 5, 4, 10],
    "Sat": [8, 15, 6, 5, 11],
    "Sun": [6, 14, 4, 3, 9]
})
df
```

```python
df["Total Sales"] = df[["Mon","Tue", "Wed","Thu","Fri","Sat","Sun"]].sum(axis=1)
print (df["Total Sales"])
```

```python
df
```

```python
best_product= df["Total Sales"].max()
print (best_product)
```

```python
df["Average"]= df [["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]].mean(axis=1)
print(df["Average"])
```

```python
print(df["Total Sales"])
```

```python
sorted = df.sort_values(by="Total Sales", ascending=False)
print(sorted)
```

```python
#exercise 4
df = pd.DataFrame(
    data = {
        "Employee": ["John", "Alice", "Bob", "Mary", "David"],
    "Mon": ["P", "P", "A", "P", "P"],
    "Tue": ["A", "P", "A", "P", "P"],
    "Wed": ["P", "P", "P", "P", "A"],
    "Thu": ["P", "P", "P", "A", "A"],
    "Fri": ["P", "A", "P", "P", "A"]
    })
df
```

```python
df["Total_Present"] = (df == "P").sum(axis=1)
df
```

```python
sorted = df.sort_values(by="Total_Present", ascending=False)
print(sorted)
```

```python
df = pd.DataFrame(
    data = {
    "Movie": ["Inception", "Titanic", "Avatar", "Joker", "Dune"],
    "Critic_A": [9, 8, 7, 9, 8],
    "Critic_B": [8, 7, 8, 9, 7],
    "Critic_C": [9, 8, 6, 10, 8]
    }
)
df
```

```python
df["Average_Rating"] = df[["Critic_A", "Critic_B", "Critic_C"]].mean(axis=1)
print(df["Average_Rating"])
```

```python
df.set_index("Movie", inplace=True)
```

```python
highest = df["Average_Rating"].idxmax()
print(highest)
```

```python
lowest = df["Average_Rating"].idxmin()
print(lowest)
```

```python
df_sorted = df.sort_values(by="Average_Rating", ascending=False)
df
```
